```python
import pandas as pd

# Función para convertir el enlace compartido de Google Drive en un enlace de descarga directa
def convertir_enlace_google_drive(enlace_compartido):
    # Extrae el ID del archivo del enlace compartido
    id_archivo = enlace_compartido.split('/')[-2]
    # Crea el enlace de descarga directa
    enlace_descarga_directa = f"https://drive.google.com/uc?id={id_archivo}"
    return enlace_descarga_directa

# Enlace compartido de Google Drive
enlace_compartido = "tu_enlace_compartido_aquí"

# Convertir el enlace compartido en un enlace de descarga directa
enlace_descarga_directa = convertir_enlace_google_drive(enlace_compartido)

# Leer el CSV con pandas
df = pd.read_csv(enlace_descarga_directa)
```
